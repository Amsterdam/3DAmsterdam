```python
import os.path

CityGMLFiles = "E:/exporttest/BuildingsLOD1/ZippedTiles/BuildingsLOD1_CityGML_{x}.0_{y}.0.zip"
CityGMLFileLink = "../data/Downloads/BuildingsLOD1_CityGML_{x}.0_{y}.0.zip"
LOD2Files = "E:/exporttest/dxf-files/LOD2/{x}-{y}-LOD2.dxf"
LOD2FileLink = "../data/Downloads/{x}-{y}-LOD2.zip"
LOD1Files = "E:/exporttest/dxf-files/LOD1/{x}-{y}-LOD1.dxf"
LOD1FileLink = "../data/Downloads/{x}-{y}-LOD1.zip"

defaultRowString = '<tr><th scope="row">{nr}</th><td>{buttons}</td><tr>'
defaultbuttonstring = '<button type="button" btn btn-primary btn-sm href="#" role="button">{buttonname} <span class="oi oi-data-transfer-download"></span></button>'

file = ""
```

```python
def GetButtonString(X,Y,FileLink,baseFileLink,buttonName):
    lod2Buttonstring=""
    LOD2File = FileLink.replace("{x}",(str)((X+xstart)*1000))
    LOD2File = LOD2File.replace("{y}",(str)((Y+ystart)*1000))
    if os.path.exists(LOD2File):
        lod2Buttonstring = defaultbuttonstring.replace("{buttonname}",buttonName)
        lod2URL = baseFileLink.replace("{x}",(str)((X+xstart)*1000))
        lod2URL = lod2URL.replace("{y}",(str)((Y+ystart)*1000))
        lod2Buttonstring = lod2Buttonstring.replace("#",lod2URL)
    return lod2Buttonstring
```

```python
def writeTAbleHeader():
    file.write('<table class="table table-striped">	<caption>Lijst met te downloaden 3D bestanden van Amsterdam in DXF LOD1, DXF LOD2 en CityGML</caption><thead><tr><th scope="col">Tegel</th><th scope="col">Bestanden</th>                </tr></thead><tbody>	')
```

```python
def WriteTableEnd():
    file.write('</tbody></table>')
```

```python
def getCellName(X,Y,aantalY):
    Xstring=chr((X)+65)
    if X>36:
        Xstring = "A"+chr((X-36)+65) 
    print(str(aantalY-Y)+Xstring)
    return str(aantalY-Y)+Xstring
    
```

```python
xstart = 109
ystart = 475

file = open("downloadtabel2.txt", "w")
writeTAbleHeader()


lod1buttonstring = ""
lod2buttonstring = ""
gmlbuttonstring = ""
aantalX = 30
aantalY = 25

for Y in reversed (range(aantalY)):
    for X in range(aantalX):
        buttonstringLOD2 = GetButtonString(X,Y,LOD2Files,LOD2FileLink, "DXF LOD2")
   
        buttonstringGML = GetButtonString(X,Y,CityGMLFiles,CityGMLFileLink, "CityGML")
 
        buttonstringLOD1 = GetButtonString(X,Y,LOD1Files,LOD1FileLink, "DXF LOD1")

        if buttonstringLOD2=="" and buttonstringLOD1 =="" and buttonstringGML =="":
            print("no tiles")
        else:
            RowName = (str)(X+1)
            columname = chr((aantalY-Y)+64)   #65 = A
            rowstring = defaultRowString.replace("{nr}",getCellName(X,Y,aantalY))
            rowstring=rowstring.replace("{buttons}",buttonstringLOD1+buttonstringLOD2+buttonstringGML)
            file.write(rowstring)
            
WriteTableEnd()
file.close()
```
